```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
data = {i: pd.read_csv(f"alpha0{i}/results/evaluation.csv") for i in [1, 2, 3, 4, 5, 6, 7, 8, 9]}
```

```python
baseline = data[1].mean()['Stage 0']
```

```python
fig, ax = plt.subplots(1)
cmap = mpl.colormaps['viridis']
for i in range(1, 10):
    ax.plot((1 - (data[i].mean() / baseline)) * 100, c=cmap(i/10), label=fr"$\alpha = 0.{i}$")
ax.set_ylabel("Percentage Improvement (%)", fontsize=12)
ax.legend(frameon=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```
